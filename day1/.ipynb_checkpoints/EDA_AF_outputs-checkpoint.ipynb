## Looking at the distogram
This is the first of many keys wthin the df1 dictionary and each will be examined.

From the paper:
We linearly project the symmetrized pair representations (zij + zji) into 64 distance bins and obtain the bin
probabilities $p^{b}_{ij}$ with a softmax. The bins cover the range from 2 Å to 22 Å; and they have equal width apart
from the last bin, which also includes any more distant residue pairs. For prediction targets $y^{b}_{ij}$ we use one-hot
encoded binned residue distances, which are computed from the ground-truth beta carbon positions for all
amino acids except glycine where use alpha carbon instead. We use the cross-entropy loss averaged over all
residue pairs:

$L_{dist}$ = -$\frac{1}{N^{2}_{}res}$ (finish this)

```python
distogram = df1['distogram']
for keys in distogram:
    print(keys)
```

The distogram has both 'bin_edges' (array) and 'logits' (3D array) as keys and arrays inside them. I will plot the data below. I understand logits to mean the conversion of (0,1) probability to real numbers (-inf - +inf).

```python
# Looking into what 'logits' contains. 
distogram['logits'].shape
```

Logits is a 3D array of (638, 638, 64) corresponding to 638 residues and 64 corresponding to the distance bins, as stated in the supplementary of the paper (https://static-content.springer.com/esm/art%3A10.1038%2Fs41586-021-03819-2/MediaObjects/41586_2021_3819_MOESM1_ESM.pdf [1.9.8 Distogram prediction])

```python
distogram['logits'][637][637].shape
```

From this exploration, it seems that there is a 64 by 64 array for each 638 by 638 residue numbers, which makes sense as the initial authors used one hot encoding to look at the predictions.

## Experimentally resolved
The model contains a head that predicts if an atom is experimentally resolved in a high-resolution structure.
The input for this head is the single representation {si} produced by the Evoformer stack (subsection 1.6). The
single representation is projected with a Linear layer and a sigmoid to atom-wise probabilities We train this head during fine-tuning (see subsubsection 1.12.1) on high-resolution X-ray crystals and cryo-EM structures (resolution better than 3Å) with standard cross-entropy

```python
experimentally_resolved = df1['experimentally_resolved']
for keys in experimentally_resolved:
    print(keys)
```

```python
experimentally_resolved['logits'].shape
```

```python
plt.plot(experimentally_resolved['logits'])
plt.xlabel('Residue number')
plt.show()
```

## This is the predicted aligned error
This is a paired residue confidence. Higher is worse.

```python
for file in files:
    df = pd.read_pickle(file) # add different models
    plt.figure(figsize=([10,8]))
    sns.heatmap(df['predicted_aligned_error'],cmap='viridis') #other colours could be cmap='inferno' or cmap='viridis'
    ptm = np.round(df['ptm'],2)
    iptm = str(np.round(df['iptm'],2))
    name = []
    file.split('_')
    name.append(file.split('_')[1])
    name.append(file.split('_')[2])
    model_name = ' '.join(name)
    file_model_name = '_'.join(name)
    plt.title(f"protein: {dir_name}, {model_name}, pTM: {ptm}, ipTM: {iptm}")
    plt.savefig(f'{dir_name}_{file_model_name}.png')
    plt.show()
```
